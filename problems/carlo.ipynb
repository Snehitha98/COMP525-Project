```python
# Import pandas - to use
import pandas as pd
import altair as alt

total_listings = pd.read_csv('output/Total_listings.csv', encoding='Latin-1')
total_listings
```

```python
total_listings_graph = alt.Chart(total_listings).mark_bar().encode(x='Number of Listings', y=alt.Y('Neighborhood Group', sort=alt.EncodingSortField(field="Number of Listings", order='descending')), color='Number of Listings', tooltip='Neighborhood Group')
total_listings_graph.save('graphs/total_listings_graph.html')
total_listings_graph
```

```python
average_price_in_neighborhood = pd.read_csv('output/average_price.csv', encoding='Latin-1')
average_price_in_neighborhood
```

```python
avg_price_graph = alt.Chart(average_price_in_neighborhood).mark_bar().encode(x='Average Price', y=alt.X('Neighborhood', sort=alt.EncodingSortField(field="Average Price", order='descending')), color='Neighborhood', tooltip='Neighborhood')
avg_price_graph.save('graphs/average_price_in_neighborhood.html')
avg_price_graph
```
